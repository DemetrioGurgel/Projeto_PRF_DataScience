**(c) quais as cinco cidades brasileiras onde mais ocorreram acidentes em rodovias federais?**

Para realizar essa questão, é preciso combinar os dados da coluna de municípios, utilizando uma função do pandas chamada de groupby, para que a partir disso criar um "contador" com a utilização do size, cuja função é calcular tamanhos de grupos, e utilizando o reset_index depois de descartar e filtrar as linhas, essa função é usada para redefinir o índice do Python DataFrame resultante.

**(b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos
considerados? (proporção = numero de ilesos ou feridos graves/ número total de
pessoas envolvidas no acidente**

As proporções de cada mês estão nos DataFrames a seguir, que estão divididos por ano. Para fazer esses dataframes, usei o datetime para converter a coluna "data_inversa" em uma data para o df, o loc para filtrar as colunas pretendidas, o depois foi usado o groupby da mesma forma das questões anteriores, somando os dados com a função .sum(), logo após dividindo as somas para realizar o cálculo de proporção. O round, logo no final, é usado para delimitar as casas decimais. 

```python
#2018
dados2018["mes"] = pd.to_datetime(dados2018["data_inversa"]).dt.month
mes_2018 = dados2018.loc[:,["mes", "pessoas", "ilesos", "feridos_graves"]]
soma = mes_2018.groupby(by="mes").sum()
prop_ilesos = (soma["ilesos"]/soma["pessoas"])
prop_feridos = (soma["feridos_graves"]/soma["pessoas"])
prop = pd.DataFrame(prop_ilesos).rename(columns={0:"proporção ilesos"}).round(2)
prop["proporção feridos graves"] = pd.DataFrame(prop_feridos).round(2)
prop
```

**(c) mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não
a incidência de acidentes nas rodovias federais.**

Para a reorganização dos dados foram feitas copias usando a função .copy(), a partir das copias foi feita uma nova coluna com a função .assing(), após isso, foi concatenado tudo em um dataframe e usado o groupby para a filtragem e contagem de quantidade de acidentes em cada ano.

```python
#2018
copia18 = dados2018.copy()
dados18 = copia18.assign(ano=2018)

#2019
copia19 = dados2019.copy()
dados19 = copia19.assign(ano=2019)

#2020
copia20 = dados2020.copy()
dados20 = copia20.assign(ano=2020)

#2021
copia21 = dados2021.copy()
dados21 = copia21.assign(ano=2021)

concatena_copia = pd.concat([dados18, dados19, dados20, dados21])
df = concatena_copia.groupby(['ano']).size().reset_index(name='acidentes')
df_concatenado = df.sort_values(by='acidentes', ascending=False)

df_concatenado
```

Analisando o dataframe acima, é possível perceber que a pandemia de certa forma conseguiu diminuir a incidência de acidentes, considerando que a pandemia começou no ano de 2020, é justamente o ano que tem menos acidentes, em 2021 tivemos um pequeno aumento, mas ainda sim temos menos acidentes do que antes da pandemia.
